Define reusable functions, work with arguments, return values, and simple checks.

8.1 Function Basics

```r
# Example: a simple function to compute GC content of a DNA string
gc_content <- function(dna) {
  # Convert to upper just in case
  s <- toupper(dna)
  n <- nchar(s)
  if (n == 0) return(NA_real_)
  gc <- sum(unlist(strsplit(s, "")) %in% c("G", "C"))
  return(100 * gc / n)
}

# Try it
gc_content("ATGCGT")
gc_content("")
```

Write a function nucleotide_counts(seq) that returns a named vector with counts of A, T, C, G.

```r
# START HERE
# Define: nucleotide_counts(seq)
nucleotide_counts <- function(seq) {
# Hints:
# - Use toupper()
seq <- toupper(seq)
# - Use strsplit(seq, "") to split into characters
seq_vec <-unlist(strsplit(seq, ""))
# - Use sum(chars == "A") etc.
count_A <- sum(seq_vec == "A")
count_T <- sum(seq_vec == "T")
count_C <- sum(seq_vec == "C")
count_G <- sum(seq_vec == "G")
# - Return c(A=..., T=..., C=..., G=...)
return(c(A = count_A, T = count_T, C = count_C, G = count_G))
}
# END HERE


print(nucleotide_counts("acgtastgtt"))
```

8.2 Arguments with Defaults & Input Validation

```r
# Example: k-mer counter with default k = 3
kmer_count <- function(seq, k = 3) {
  s <- toupper(seq)
  n <- nchar(s)
  if (!is.numeric(k) || length(k) != 1 || k < 1) {
    stop("k must be a positive integer length 1.")
  }
  if (k > n) return(integer(0))
  kmers <- sapply(1:(n - k + 1), function(i) substr(s, i, i + k - 1))
  tab <- table(kmers)
  return(tab[order(names(tab))])
}

# Try it
kmer_count("ATGCGT", k = 2)
```

Write a function safe_mean(x, na_rm = TRUE) that:

* checks x is numeric,

* if na_rm is TRUE, ignores NAs,

* returns NA if all values are NA or x is length 0.

```r
# START HERE
# Define: safe_mean(x, na_rm = TRUE)
# Hints:
safe_mean <- function(x, na_rm = TRUE) {
 if (!is.numeric(x)){
  stop("x must be numeric")
 }
# - if (length(x) == 0) return(NA_real_)
if ((length(x)) == 0 || all(is.na(x))){
      return(NA_real_)
}
# - mean(x, na.rm = na_rm)
 mean(x, na.rm = na_rm)
}

# END HERE
```

8.3 Returning Multiple Values (Lists) & Simple Tests

```r
# Example: summary of a numeric vector
vec_summary <- function(x) {
  if (!is.numeric(x)) stop("x must be numeric.")
  if (length(x) == 0) return(list(mean = NA_real_, sd = NA_real_, n = 0))
  out <- list(
    mean = mean(x, na.rm = TRUE),
    sd   = sd(x, na.rm = TRUE),
    n    = sum(!is.na(x))
  )
  return(out)
}

vec_summary(c(1, 2, 3, NA))
```

# **Basic Statistics: t-test & Multiple Testing (BH/FDR)**

run simple tests and control FDR (like BH in RNA-seq workflows).

```r
set.seed(42)
grp <- rep(c("treated","control"), each = 10)
x   <- rnorm(20, mean = ifelse(grp=="treated", 1, 0), sd = 1)

# Two-sample t-test
t.test(x[grp=="treated"], x[grp=="control"])
```

```r
# START HERE
# Simulate 100 genes: for each gene, generate 10 treated and 10 control values (normal).
number_of_genes <- 100
num_samples_per_group <-10
data_matrix <- matrix(NA, nrow = number_of_genes  , ncol =  2 * num_samples_per_group)
mean_shift <- c(rep(1, 50), rep(0, 50))
# For half the genes, add a mean shift of +1 to treated to simulate DE.
for (i in 1:number_of_genes) {
  control_values <- rnorm(num_samples_per_group, mean = 0, sd = 1)
  treated_values <- rnorm(num_samples_per_group, mean = mean_shift[i], sd = 1)
  data_matrix[i, ] <- c(treated_values, control_values)
}
# 1) For each gene, run a t-test and collect p-values.
p_value_vector <- numeric(number_of_genes)

treated_cols <- 1:10
control_cols <- 11:20

for (i in 1:number_of_genes) {
  test_result <- t.test(data_matrix[i, treated_cols], data_matrix[i, control_cols])
  p_value_vector[i] <- test_result$p.value
}
# 2) Adjust p-values with BH (p.adjust(..., method="BH")).
adjusted_p_values <- p.adjust(p_value_vector, method = "BH")
# 3) Count how many genes are significant at FDR < 0.05.
significant_genes_count <- sum(adjusted_p_values < 0.05)
print(significant_genes_count)



# END HERE
```

# **Correlation, Heatmap, and Clustering**

quick QC / similarity checks between samples.

```r
# Build a small expression matrix (genes x samples)
set.seed(1)
mat <- matrix(rnorm(200), nrow = 20)  # 20 genes, 10 samples
print(mat)
colnames(mat) <- paste0("S", 1:10)

# Correlation between samples
cors <- cor(mat)
# Heatmap (base R)
heatmap(cors)
```

```r
# START HERE
# 1) Scale each gene (row) to mean 0 / sd 1 -> use scale(t(mat)) trick or apply.
scaled_mat <- t(scale(t(mat)))
# 2) Compute correlation of the (scaled) matrix across samples.
sample_cors <- cor(scaled_mat)
# 3) Perform hierarchical clustering: dist() + hclust(), and plot the dendrogram.
sample_distances <- dist(t(scaled_mat), method = "euclidean")
sample_hclust <- hclust(sample_distances, method = "complete")
plot(sample_hclust, main = "Hierarchical Clustering of Samples",
     xlab = "Samples (S1-S10)", sub = "")


# END HERE
```
